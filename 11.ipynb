```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import sympy as sp
import random

max_value=0.5
min_value=0

seismic = np.loadtxt(
    '/Users/lihaoyang/研究生/研一/课题组/数据集/lbm_data/lbm_Seismic.csv',
    dtype=float, delimiter=',')
impedance = np.loadtxt(
      '/Users/lihaoyang/研究生/研一/课题组/数据集/lbm_data/lbm_Imp.csv',
    dtype=float, delimiter=',')
time = np.loadtxt(
      '/Users/lihaoyang/研究生/研一/课题组/数据集/lbm_data/lbm_time.csv',
    dtype=float, delimiter=',')
seismic = np.array(seismic).T
impedance = np.array(impedance).T
time = np.array(time).T
seismic = max_value*2* ((seismic - seismic.min()) / (seismic.max() - seismic.min())) + min_value
impedance = max_value*2* ((impedance - impedance.min()) / (impedance.max() - impedance.min())) + min_value

time =  ((time - time.min(axis=0)) / (time.max(axis=0) - time.min(axis=0)))
seismic = seismic.T
impedance = impedance.T
time = time.T

plt.title("impedance")
plt.imshow(impedance.T,vmin=0,vmax=1)
cbar = plt.colorbar(label='Difference', fraction=0.02, pad=0.04)
```
