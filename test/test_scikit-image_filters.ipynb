使用scikit-image库进行边缘检测

```python
import matplotlib.pyplot as plt
from torchvision.datasets import MNIST
from torchvision import transforms
from skimage import filters

# 加载MNIST数据集
transform = transforms.Compose([
    transforms.ToTensor(),
])
mnist_data = MNIST(root='../data/img', train=True, download=True, transform=transform)

# 获取指定索引的图片
indices = [1999, 213, 3456, 92]
images = [mnist_data[i][0].unsqueeze(0) for i in indices]
labels = [mnist_data[i][1] for i in indices]

# 将PyTorch张量转换为NumPy数组
images_np = [image.numpy() for image in images]

# 应用Prewitt算子的水平和垂直方向
prewitt_h_edges = [filters.prewitt(image, axis=0) for image in images_np]
prewitt_w_edges = [filters.prewitt(image, axis=1) for image in images_np]
prewitt_edges = [filters.prewitt(image) for image in images_np]

# 显示原始图像和Prewitt边缘检测后的图像
fig, axs = plt.subplots(4, 4, figsize=(12, 12))

for i, (image, label, h_edge, w_edge, p_edge) in enumerate(zip(images, labels, prewitt_h_edges, prewitt_w_edges, prewitt_edges)):
    axs[i, 0].imshow(image[0, 0], cmap='viridis')
    axs[i, 0].set_title('MNIST')
    axs[i, 0].axis('off')

    axs[i, 1].imshow(h_edge[0, 0], cmap='viridis')
    axs[i, 1].set_title('Prewitt_x')
    axs[i, 1].axis('off')

    axs[i, 2].imshow(w_edge[0, 0], cmap='viridis')
    axs[i, 2].set_title('Prewitt_y')
    axs[i, 2].axis('off')

    axs[i, 3].imshow(p_edge[0, 0], cmap='viridis')
    axs[i, 3].set_title('Prewitt')
    axs[i, 3].axis('off')

plt.tight_layout()
plt.show()
```
